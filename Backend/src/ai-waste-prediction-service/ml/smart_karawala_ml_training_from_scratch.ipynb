```python
import pandas as pd
import joblib
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
waste_data = pd.read_csv('dataset.csv')
waste_data.head()
```

```python
X_waste = waste_data[['Fish Type', 'Raw weight']]
y_waste = waste_data['Waste Amount']

X_train, X_test, y_train, y_test = train_test_split(
    X_waste,
    y_waste,
    test_size=0.2,
    random_state=42
)

waste_model = Pipeline([
    ('encoder',
        ColumnTransformer([
            ('Fish Type',
             OneHotEncoder(handle_unknown='ignore'),
             ['Fish Type'])
        ], remainder='passthrough')
    ),

    ('regressor', LinearRegression())
])

waste_model.fit(X_train, y_train)

print("Waste model trained successfully")
```

```python
y_pred = waste_model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print('Waste Model Evaluation')
print('MAE:', round(mae, 4))
print('RMSE:', round(rmse, 4))
print('R2 Score:', round(r2, 4))
```

```python
joblib.dump(waste_model, 'waste_model.pkl')
print('Waste model saved as waste_model.pkl')
```

```python
test_waste = pd.DataFrame({'Fish Type': ['Sprats'], 'Raw weight': [50]})
print('Predicted Waste:', round(float(waste_model.predict(test_waste)[0]), 2), 'kg')
```

```python
salt_data = pd.read_csv('salt_dataset.csv')
salt_data.head()
```

```python
X_salt = salt_data[['fish_type', 'cleaned_weight']]
y_salt = salt_data['salt_amount']

X_train_s, X_test_s, y_train_s, y_test_s = train_test_split(X_salt, y_salt, test_size=0.2, random_state=42)

salt_model = Pipeline([
    ('encoder', ColumnTransformer([
        ('fish_type', OneHotEncoder(handle_unknown='ignore'), ['fish_type'])
    ], remainder='passthrough')),
    ('regressor', LinearRegression())
])

salt_model.fit(X_train_s, y_train_s)
print('Salt model trained successfully')
```

```python
y_pred_s = salt_model.predict(X_test_s)

mae_s = mean_absolute_error(y_test_s, y_pred_s)
mse_s = mean_squared_error(y_test_s, y_pred_s)
rmse_s = np.sqrt(mse_s)
r2_s = r2_score(y_test_s, y_pred_s)

print('Salt Model Evaluation')
print('MAE:', round(mae_s, 4))
print('RMSE:', round(rmse_s, 4))
print('R2 Score:', round(r2_s, 4))
```

```python
joblib.dump(salt_model, 'salt_model.pkl')
print('Salt model saved as salt_model.pkl')
```

```python

test_salt = pd.DataFrame({'fish_type': ['Sprats'], 'cleaned_weight': [40]})
print('Predicted Salt:', round(float(salt_model.predict(test_salt)[0]), 2), 'kg')
```
